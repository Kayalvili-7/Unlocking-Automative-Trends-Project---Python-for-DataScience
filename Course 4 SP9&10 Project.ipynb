```python
 def impute_missing_values(Car_df):
    for col in Car_df.columns:
        if Car_df[col].dtype == 'object':  # Categorical column
            Car_df[col].fillna('Not Available', inplace=True)
        else:# Numerical column
            Car_df[col].fillna(Car_df[col].mean(), inplace=True)
    return Car_df
```

```python
#Task 1

# Define safety features
safety_features = ['ABS_(Anti-lock_Braking_System)', 'Airbags', 'Hill_Assist']

# Create binary columns for safety features
for feature in safety_features:
    data[feature] = data[feature].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculate feature presence percentage
feature_presence = data[safety_features].sum() / len(data) * 100

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=feature_presence.index, y=feature_presence.values)
plt.xlabel('Safety Feature')
plt.ylabel('Percentage of Cars')
plt.title('Presence of Safety Features in Car Models')
plt.show()

#INTERPRETATION: The chart shows the percentage of car models that come equipped with Anti-lock Braking System (ABS),
#Airbags, and Hill Assist. Airbags appear to be the most common safety feature, followed by ABS and then Hill Assist.
```
